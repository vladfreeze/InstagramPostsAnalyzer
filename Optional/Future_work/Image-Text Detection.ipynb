```python



# Function to extract words from an image URL using OCR
def extract_words_from_image(url):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open("temp_image.jpg", "wb") as f:
            f.write(response.content)
        image = Image.open("temp_image.jpg")
        if image.format == "JPEG":
            text = pytesseract.image_to_string(image)
            return text
    return None

```

```python
# Apply the function to each post data
for post in post_data:
    post_id = post['id']
    media_url = post['media_url']
    extracted_text = extract_words_from_image(media_url)
    print(extracted_text)
    df.loc[df['id'] == post_id, 'extracted_text'] = extracted_text

# Print the DataFrame with extracted text
print(df)
```

```python

def split_caption_column(caption):
    actual_text_column = []
    hashtags_column = []

   
    for caption in caption:
        try:
            words = caption.split()

            actual_text = []
            hashtags = []

            for word in words:
                if word.startswith('#'):
                    hashtags.append(word)
                else:
                    actual_text.append(word)

            actual_text = ' '.join(actual_text)
            hashtags = ' '.join(hashtags)

            actual_text_column.append(actual_text)
            hashtags_column.append(hashtags)
        except Exception as e:
            print(caption)
            
            print(e)
    return actual_text_column, hashtags_column

actual_text_column, hashtags_column = split_caption_column(df['caption'])

# Add the new columns to the DataFrame
df['actual_text'] = actual_text_column
df['hashtags'] = hashtags_column
```
